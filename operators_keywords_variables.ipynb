```python
'''
Ⅰ) Arthmetic Operators
    ⇒ Python offers arithmetic operators for executing mathematical operations on variables, numbers, or expressions.
    ⇒ They enable calculations such as addition, subtraction, multiplication, division, modulus, exponentiation, and floor division.

Emplementation
'''
# Addition 
x = 3 + 4 # x = 7

# Subtraction
y = 8 - 5  # y is 3

# Multiplication
z = 2 * 6  # z is 12

# Division
a = 10 / 2  # a is 5.0

# Modulo
b = 10 % 3  # b is 1

# Exponentiation
c = 2 ** 3  # c is 8

# Floor Division
d = 10 // 3  # d is 3

print(x, "\n", y, "\n", z, "\n", a, "\n", b, "\n", c, "\n", d)
```

```python
'''
Ⅱ) Comparision Operators
    ⇒ Comparison operators are used to compare the values of two variables and identify their relationship.
    ⇒ These python operators enable conditional decision-making in programs by returning a boolean value (True or False) based on the comparison.
    ⇒ They can do tests, for equality, inequality, greater than, less than, greater than or equal to, and less than or equal to.

Emplementation
'''
# Equal to
x = 3 == 3 # x is True

# Not equal to
y = 4 != 2 # y is True

# Greater than
z = 9 > 6 # z is True

# Less than
a = 5 < 8 # a is True

# Greater than  or equal to 
b = 6 >= 9 # b is True

# Less then or equal to 
c = 2 <= 3 # c is True

print(x, "\n", y, "\n", z, "\n", a, "\n", b, "\n", c)
```

```python
'''
Ⅲ) Logical Operators
    ⇒ In conditional statements, Python logical operators are used to combine multiple conditions and make conclusions depending on the truth value of the conditions.
    ⇒ They are required for integrating numerous conditions in control structures such as if statements and while loops. AND, OR, and NOT are the three fundamental logical operators.

Emplementation
'''
# and
x = (3 > 2) and (4 < 5) # x is true
# and operator returns true if and only if both or operands are true.

# or
y = (2 == 2) or (3 != 3) # y is true
# or operator returns true if either one of its condition is true.

# not
z = not (2 < 1) # z is true
# not operator negates a boolean expression. It returns the opposite of the boolean value of its operand.

print(x, "\n", y, "\n", z)
```

```python
'''
Ⅳ) Assignment Operators
    ⇒ Variables were given values using assignment operators.
'''
# Assignment operator (=)
a = 5
print(a)
# assigns the value 5 to the variable "a".

# Compound assignment operator (+=)
a += 5 # a = a + 5
print(a)
# adds 5 to the current value of "a" and assigns the result back to "a". So now "a" is equal to 10.

# Compound assignment operator (-=)
a -= 5 # a = a - 5
print(a)
# subtracts 5 from the current value of "a" and assigns the result back to "a". So now "a" is equal to 5.

# Compound assignment operator (*=)
a *= 5 # a = a * 5
print(a)
# multiplies the current value of "a" by 5 and assigns the result back to "a". So now "a" is equal to 25.

# Compound assignment operator (/=)
a /= 5 # a = a / 5
print(a)
# divides the current value of "a" by 5 and assigns the result back to "a". So now "a" is equal to 5.

# Compound assignment operator (%=)
a %= 5 # a = a % 5
print(a)
# calculates the modulo of the current value of "a" with 5 and assigns the result back to "a". So now "a" is equal to 0.
```

```python
'''
Ⅴ) Identity Operators
    ⇒ Identity operators in Python are used to determine whether or not two objects have the same memory addresses. This differs from comparison operators, which compare object values rather than memory locations.
'''
x = 5
y = 5
z = 10

# The 'is' operator outputs True when two variables relate to the same object in memory.
print(x is y)  # Output: True
print(x is z)  # Output: False

# The 'is not' operator outputs True when the variables do not relate to the same object.
print(x is not z)  # Output: True

print("id(x) = ", id(x))
print("id(y) = ", id(y))
print("id(z) = ", id(z))

a: list = [1, 2, 3]
b: list = [1, 2, 3]
c: list = a

print(a is c)       # True  (same object, sharing same memmory space)
print(a is b)       # False (different objects, seperate memmory space)
print(a == b)      # True  (same values, different memmory space but having same vlaues)
print(a is not b)  # True  (True because of different memmory space, although having same memmory space)

print('\n-----\n') # seperator for better output readability

print("id(a) = ", id(a))
print("id(b) = ", id(b))
print("id(c) = ", id(c))
```

```python
'''
Ⅵ) Membership Operator
    ⇒ Membership operators identify whether a variable is a member of a sequence, such as strings, lists, or tuples. or not.
    ⇒ They are helpful to find out if a variable in a data structure exists or not. If the value provided is found in the series, the 'in' operator returns True, and if it is not, the 'not in' operator returns True.
'''
# in
_list: list = [1, 2, 3, 4, 5]
res = 3 in _list # res is true
# The in operator returns True if the specified value or object is found in the sequence and False otherwise.

# not in
_names = ["Ali", "Sana", "Sara", "zain"]
res = "Sikandar" not in _names # res is true cause Sikandar is not in my list
# The not in operator returns True if the specified value or object is not found in the sequence and False otherwise.

# String
student_name = 'Sikandar Shehriyar'
res = "w" in student_name # returns false
res_2 = "w" not in student_name # returns true

print(res, '\n', res_2)
```
